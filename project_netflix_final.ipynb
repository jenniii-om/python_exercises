Perform exploratory data analysis on the `netflix_data.csv` data to understand more about movies from the 1990s decade.

- What was the most frequent **<u>movie</u>** duration in the <u>1990s</u>? Save an approximate answer as an integer called duration.

- A movie is considered short if it is less than 90 minutes. Count the number of **<u>short action movies</u>** released in the 1990s and save this integer as short_movie_count.

```python
# Importing pandas and matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("csv_files/netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
release_yr_start= 1990
release_yr_end = 1999
show_type = 'Movie'

sub = np.logical_and(np.logical_and(netflix_df['release_year'] >= release_yr_start, netflix_df['release_year'] <= release_yr_end),
                      netflix_df['type'].str.lower() == show_type.lower())

netflix_sub_df = netflix_df[sub]

#print(type(netflix_sub_df))
#print(netflix_sub_df)

duration = netflix_sub_df.value_counts('duration', ascending=False).index[0]
#print(type(duration))
#print(duration)

print(f"Frequent movie duration in {release_yr_start}s: {duration}")
```

```python
dur = 90
genre = 'Action'

# using numpy logical and operator to filter out movies with duration less than the variable dur, is within the release year, and equals to the genre variable
short_movie = np.logical_and(netflix_sub_df['duration'] < dur, netflix_df['genre'].str.lower() == genre.lower())
short_movie_count = len(netflix_sub_df[short_movie])

print(f"Short {genre} movies in {release_yr_start}s: {short_movie_count}")

```
